# Z-Score

```python
adr_mean=df.adr.mean()
```

```python
adr_std=df.adr.std()
```

```python
print(adr_mean,adr_std)
```

```python
df['zscore']=(df.adr-adr_mean)/adr_std
```

```python
df.head()
```

# All_Outliers Away From 3rd Standard Deviation

```python
outliers=df[(df.zscore < -3) | (df.zscore > 3)]
```

```python
outliers.shape
```

```python
outliers.head()
```

# Removing Outlier

```python
df.shape
```

```python
df=df[(df.zscore >= -3) & (df.zscore <= 3)]
```

# Data Analysis and Visualization

```python
canceled_perc = df.is_canceled.value_counts(normalize = True)
canceled_perc
```

```python
sns.set_style('darkgrid')
a=sns.countplot(x=df.is_canceled,palette='Blues',edgecolor='k')
plt.title('Reservation status count',size=18)
plt.xticks([0,1],['Not Canceled','Canceled'])
for i in a.containers:
    a.bar_label(i)
plt.show()
```

```python
sns.countplot(x= 'hotel', data=df,hue='is_canceled',palette='GnBu',edgecolor='k')
plt.title('Reservation status in diffrent hotel',size=18)
plt.legend(['Not Canceled','Canceled'])
plt.ylabel('No. of reservations',size=12)
plt.show()
```

```python
resort_hotel = df[df.hotel == 'Resort Hotel']
resort_hotel.is_canceled.value_counts(normalize=True)
```

```python
city_hotel = df[df.hotel == 'City Hotel']
city_hotel.is_canceled.value_counts(normalize=True)
```

```python
resort_hotel=resort_hotel.groupby('reservation_status_date')[['adr']].mean().reset_index()
city_hotel=city_hotel.groupby('reservation_status_date')[['adr']].mean().reset_index()
```

```python
plt.figure(figsize=(20,8))
sns.lineplot(x='reservation_status_date',y='adr',data=resort_hotel,label='Resort Hotel')
sns.lineplot(x='reservation_status_date',y='adr',data=city_hotel,label='City Hotel')
plt.title('Average Daily Rate in City and Resort Hotel',size=25)
plt.xlabel('reservation status date',size=18)
plt.ylabel('average daily rate',size=18)
plt.legend(fontsize=15)
plt.show()
```

```python
plt.figure(figsize=(20,8))
df['month']=df['reservation_status_date'].dt.month
sns.countplot(x='month',data=df,hue='is_canceled',palette='GnBu',edgecolor='k')
plt.title('Monthly Reservations',size=25)
plt.xlabel('month',size=18)
plt.ylabel('number of reservations',size=18)
plt.legend(['Not Canceled','Canceled'],fontsize=15)
plt.show()
```

```python
plt.figure(figsize=(12,5))
sns.barplot(x='month',y='adr',data=df[df['is_canceled']==1].groupby('month')[['adr']].sum().reset_index())
plt.title('Adr for each month',size=20)
plt.xlabel('month',size=15)
plt.ylabel('average daily rate',size=15)
plt.show()
```

```python
plt.figure(figsize=(6,6))
canceled_data = df[df['is_canceled']==1]
top_10_country = canceled_data['country'].value_counts().head(10)
plt.title('Top 10 Country with reservation Canceled',size=15)
plt.pie(top_10_country,autopct='%.2f%%',labels=top_10_country.index)
plt.show()
```

```python
plt.figure(figsize=(20,7))
canceled_adr = canceled_data.groupby('reservation_status_date')[['adr']].mean()
canceled_adr.reset_index(inplace=True)

not_canceled_data = df[df['is_canceled']==0]
not_canceled_adr = not_canceled_data.groupby('reservation_status_date')[['adr']].mean().reset_index()

sns.lineplot(x='reservation_status_date',y='adr',data=canceled_adr,label='canceled')
sns.lineplot(x='reservation_status_date',y='adr',data=not_canceled_adr,label='not_canceled')
plt.title('Average Daily Rate Based on Cancellation',size=25)
plt.xlabel('reservation status date',size=18)
plt.ylabel('average daily rate',size=18)
plt.legend(fontsize=18)
plt.show()
```
